```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the second wine dataset
wine2 = pd.read_csv("winemag-data-130k-v2.csv")
```

```python
wine2.head(10)
```

```python
# Check missing values
print(wine2.isna().sum().sort_values())
```

```python
# Drop all NAs in price and country in wine2 dataset
wine2= wine2.dropna(subset=["country","price"])
print(wine2.isna().sum().sort_values())
```

```python
# Using points and numerical and country as categorical
# Let's investigate country variable
# List all countries`
list(wine2['country'].value_counts().index)
```

```python
# Count the number of countries
wine2['country'].value_counts().count()
```

```python
# Count the total number of values of each country
wine2['country'].value_counts()
```

```python
# Define countries interested, top 10 countries
country2 = ["US", "France", "Italy", "Spain", "Portugal", "Chile", "Argentina", "Austria", "Australia", "Germany"]
```

```python
# Get rows of 10 top countries
wine2a = wine2.loc[wine2['country'].isin(country2)]
```

```python
wine2a['country'].value_counts()
```

```python
wine2a.info()
```

```python
# Get country, points, and price
wine2a = wine2a.loc[:, ["country","points","price"]]
```

```python
wine2a.info()
```

```python
# Creating Response and Feature variables
Xcountry = wine2a.loc[:, ["points", "country"]]
y = wine2a.loc[:, ["price"]]
```

```python
Xcountry.head(10)
```

```python
# Reset index of X to start from 0
Xcountry = Xcountry.reset_index(drop=True)
```

```python
# Encoding dummy variables for country using OneHotEncoder and normalizing numeric variable using Standard Scaler
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
# Create instance of one-hot-encoder, dropping the first dummy variable variable
enc = OneHotEncoder(drop='first') 
# coverting categorical variable to binary variables of 1 and 0
```

```python
enc_df = pd.DataFrame(enc.fit_transform(Xcountry[['country']]).toarray())
```

```python
enc_df.head(10)
# Which columns correspond to which country
# When labels are sorted alphabetically, the first, in this case Argentina will be the baseline, which means it will contain all zeroz
# Followed by Australia, Austria, Chile, France, Germany, Italy, Portugal, Spain, US
```

```python
# Naming columns
enc_df.columns = ['Australia', 'Austria', 'Chile', 'France', 'Germany', 'Italy', 'Portugal', 'Spain', 'US']
```

```python
enc_df.head(10)
```

```python
num = StandardScaler()

# What does normalizing do?

# Substract each observation by mean and divide by the standard deviation, SD
# When numeric variable is normalized, we are expressing coefficients as the effect of a one standard deviation change in X
# Interpret as a change in Y measured for one SD change in X

# Why do we have to normalize numerical feature?

# 1. Avoid bias - Some algorithms give more weight to features with larger scales i.e variables with larger values as more important
# 2. Facilitating convergence - Rescaling numerical data help algorithms converge more quickly and find optimal solutions more efficiently
# 3. Stability of Numerical computations - Prevent numerical instabiliy issues
# 4. Interpretability - Normalization does not affect relationship between features and target variable and interpretability of the model's results
# 5. Compatibility with certain algorithms - Normalizing numerical data ensures fair comparisons and accurate results. Some algorithms such as support vector machines (SVMs) 
# or k-nearest neighbors are influenced by the scale of the features
```

```python
num_df = pd.DataFrame(num.fit_transform(Xcountry[['points']]))
```

```python
# Naming columns
num_df.columns = ["points_normalized"]
```

```python
num_df.head(10)
```

```python
# Merge num_df and enc_df`
Xcountry = pd.merge(num_df, enc_df, left_index=True, right_index=True)
```

```python
Xcountry.head(10)
```

```python
# Split data into training and test sets
# First, import relevant packages
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xcountry, y, test_size=0.30, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Instantiate LinearRegression object and fit the model
reg = LinearRegression()
model = reg.fit(X_train, y_train)
```

```python
# Evaluation of model performance
# Getting coefficients and intercept
model.coef_

# When numeric variable is normalized, we are expressing coefficients as the effect of a one standard deviation change in X

# Coefficient results when points is not normalized and used directly in model
# array([[  5.6395784 ,   1.15756242, -13.28753798,  -2.47938646, 4.9612925 ,  -0.47158683,   3.70452558,  -7.57185606, 0.09758379,   1.20432032]])``
```

```python
model.intercept_
```

```python
# Predicting y using model
y_pred = reg.predict(X_test)
```

```python
# Evaluate model performance
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print("r2: ",r2)

# R2 measures how much irregularity in the dependent variable the model can explain
# Bigger value shows a better fit between prediction and actual value
# score of 0.187 is pretty similar to r2 obtained with just points as feature variable
```

```python
# Calculate rmse
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("rmse: ",rmse)

# RSE: measure of the typical size of the residuals, how much the predictions are typically wrong
# rse of 37 means that the difference between the predicted value of price and actual price is typically about ~37
# The model has an average error for price of around $6 per point
```

```python
# Calculate adjusted R-squared (with points and country variables)

n = len(y)
R2 = r2
p = 11
```

```python
Adj_r2 = 1-(1-R2)*(n-1)/(n-p-1)
print("Adj rsquared: ", Adj_r2)

# Unlike the R2 statistic, the adjusted R2 statistic pays a price for the inclusion of unnecessary variables in the model
# Adjusted R-squared increases only when independent variable is significant and affects dependent variable. 
```
