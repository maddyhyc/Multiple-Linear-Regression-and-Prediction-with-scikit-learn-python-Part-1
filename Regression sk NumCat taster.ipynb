```python
# Define tasters interested, top 10 tasters
taster2 = ["Roger Voss", "Michael Schachner", "Kerin O’Keefe", "Virginie Boone", "Paul Gregutt", "Matt Kettmann", "Joe Czerwinski", "Sean P. Sullivan", "Anna Lee C. Iijima", "Jim Gordon"]
```

```python
# Get price, points, and taster_name
wine2b = wine2b.loc[:, ["price","points","taster_name"]]
```

```python
wine2b.info()
```

```python
# Creating Response and Feature variables
X = wine2b.loc[:, ["points", "taster_name"]]
y = wine2b.loc[:, ["price"]]
```

```python
X.head(10)
```

```python
# Reset index of X to start from 0
X = X.reset_index(drop=True)
```

```python
# Encoding dummy variables for taster_name using OneHotEncoder
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
# Create instance of one-hot-encoder, dropping the first dummy variable variable
enc = OneHotEncoder(drop='first') 
```

```python
enc_df2 = pd.DataFrame(enc.fit_transform(X[['taster_name']]).toarray())
```

```python
enc_df2.head(10)
# Which columns correspond to which taster_name
# When labels are sorted alphabetically, the first, in this case Anna Lee C. Iijima will be the baseline, which means it will contain all 0
# Followed by Jim Gordon, Joe Czerwinski, Kerin O'Keefe, Matt Kettmann, Michael Schachner, Paul Gregutt, Roger Voss, Sean P. Sullivan, Virginie Boone
```

```python
# Naming columns
enc_df2.columns = ["Jim Gordon", "Joe Czerwinski", "Kerin O'Keefe", "Matt Kettmann", "Michael Schachner", "Paul Gregutt", "Roger Voss", "Sean P. Sullivan", "Virginie Boone"]
```

```python
enc_df2.head(10)
```

```python
num = StandardScaler()

# What does normalizing do?

# Substract each observation by mean and divide by the standard deviation, SD
# When numeric variable is normalized, we are expressing coefficients as the effect of a one standard deviation change in X
# Interpret as a change in Y measured for one SD change in X

# Why do we have to normalize numerical feature?

# 1. Avoid bias - Some algorithms give more weight to features with larger scales i.e variables with larger values as more important
# 2. Facilitating convergence - Rescaling numerical data help algorithms converge more quickly and find optimal solutions more efficiently
# 3. Stability of Numerical computations - Prevent numerical instabiliy issues
# 4. Interpretability - Normalizatioin does not affect relationship between features and target variable and interpretability of the model's results
# 5. Compatibilith with certain algorithms - Normalizing numerical data ensures fair comparisons and accurate results. Some algorithms such as support vector machines (SVMs) 
# or k-nearest neighbors are influenced by the scale of the features
```

```python
num_df2 = pd.DataFrame(num.fit_transform(X[['points']]))
```

```python
# Naming columns
num_df2.columns = ["points_normalized"]
```

```python
num_df2.head(10)
```

```python
# Merge num_df2 and enc_df2
Xtaster = pd.merge(num_df2, enc_df2, left_index=True, right_index=True)
```

```python
Xtaster.head(10)
```

```python
# Split data into training and test sets
# First, import relevant packages
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xtaster, y, test_size=0.30, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Instantiate LinearRegression object and fit the model
reg = LinearRegression()
model = reg.fit(X_train, y_train)
```

```python
# Getting coefficients and intercept
model.coef_

# When numeric variable is normalized, we are expressing coefficients as the effect of a one standard deviation change in X
# if points changes by one standard deviation, then price goes up by ~$18
```

```python
print(np.std(X['points']))
```

```python
model.intercept_
```

```python
# Predicting y using model
y_pred = reg.predict(X_test)
```

```python
# Evaluate model performance
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print("r2: ",r2)

# R2 measures how much irregularity in the dependent variable the model can explain
# Bigger value shows a better fit between prediction and actual value
# score of 0.154 is pretty similar to r2 obtained with just points as feature variable
```

```python
# Calculate rmse
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("rmse: ",rmse)

# RSE: measure of the typical size of the residuals, how much the predictions are typically wrong
# rse of ~45 means that the difference between the predicted value of price and actual price is typically about ~45
# The model has an average error for price of around $45 per point
```

```python
# Calculate adjusted R-squared (using points and taster_name variables)

n = len(y)
R2 = r2
p = 11
```

```python
Adj_r2 = 1-(1-R2)*(n-1)/(n-p-1)
print("Adj rsquared: ", Adj_r2)

# Unlike the R2 statistic, the adjusted R2 statistic pays a price for the inclusion of unnecessary variables in the model
# Adjusted R-squared increases only when independent variable is significant and affects dependent variable.
```
